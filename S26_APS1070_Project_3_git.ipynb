```python
import pandas as pd
import requests, io

headers = {"User-Agent": "Mozilla/5.0"}
raw_data = pd.read_csv(
    filepath_or_buffer=io.StringIO(requests.get(
        'https://ourworldindata.org/grapher/population.csv?v=1&csvType=full&useColumnShortNames=true',
        headers=headers).text),
    index_col=False)

countries_list_url = "https://raw.githubusercontent.com/Sabaae/Dataset/main/countries_list.txt"
```

```python
### YOUR CODE HERE ###

# building countries list
txt = requests.get(countries_list_url, headers=headers).text
countries = [c.strip() for c in txt.split("\n") if c.strip()]
print(f"Countries in list: {len(countries)}")

# reshaping population data
data_filtered = raw_data[raw_data["year"].between(1901, 2023)]
df_wide = data_filtered.pivot(index="entity", columns="year", values="population_historical")
population_df = (df_wide[df_wide.index.isin(countries)]).dropna(how="any")

print(f"final population_df shape: {population_df.shape}")
population_df.head(10)
```

```python
### YOUR CODE HERE ###

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

def time_series(df, country_list):
    n = len(country_list)
    fig, axs = plt.subplots(n, 1, figsize=(10, 3*n))
    if n == 1:
        axs = [axs]
    for i, country in enumerate(country_list):
        axs[i].plot(df.columns, df.loc[country])
        axs[i].set_title(country)
        axs[i].set_ylabel("Population")
        if i == n - 1:
            axs[i].set_xlabel("Year")

    plt.tight_layout()
    plt.show()

time_series(population_df, ["Japan", "Latvia", "Austria"])

# standardization
scaler = StandardScaler()
df_std = pd.DataFrame(
    scaler.fit_transform(population_df),
    columns=population_df.columns,
    index=population_df.index
)

# string edit distance
!pip install Levenshtein
import Levenshtein

distances = {}
for c in df_std.index:
    distances[c] = Levenshtein.distance(c, "Reza")
distances = pd.Series(distances).sort_values(kind="stable")
ties = distances[distances == distances.min()].sort_index()
print(f"\nclosest countries to my name are:\n{ties.head(2)}")

```

```python
p4_countries = ["Japan", "Bosnia and Herzegovina", "Latvia", "United Arab Emirates",
                ties.index[0], ties.index[1]]

def compare_raw_std(df_raw, df_std, country_list):
    n = len(country_list)
    fig, axs = plt.subplots(n, 2, figsize=(14, 3*n))
    for i, country in enumerate(country_list):
        axs[i, 0].plot(df_raw.columns, df_raw.loc[country])
        axs[i, 0].set_title(f"{country} — raw")
        axs[i, 0].set_ylabel("Population")

        axs[i, 1].plot(df_std.columns, df_std.loc[country])
        axs[i, 1].set_title(f"{country} — standardized")
        axs[i, 1].set_ylabel("Std value")

        if i == n - 1:
            axs[i, 0].set_xlabel("Year")
            axs[i, 1].set_xlabel("Year")
    plt.tight_layout()
    plt.show()

compare_raw_std(population_df, df_std, p4_countries)
```
